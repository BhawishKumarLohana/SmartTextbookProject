```python
import os
import subprocess
import sys

# Define paths
base_dir = r"C:\Users\User\Desktop\Language Centre"
names_file = os.path.join(base_dir, "names.txt")
venv_dir = os.path.join(base_dir, "env")
venv_python = os.path.join(venv_dir, "Scripts", "python.exe")
marker_single_exe = os.path.join(venv_dir, "Scripts", "marker_single.exe")

# Step 1: Create a virtual environment
def create_virtual_env():
    try:
        if not os.path.exists(venv_dir):
            print("Creating virtual environment...")
            subprocess.run([sys.executable, "-m", "venv", venv_dir], check=True)
            print("Virtual environment created at:", venv_dir)
        else:
            print("Virtual environment already exists.")
    except subprocess.CalledProcessError as e:
        print("Error creating virtual environment:", e)

# Step 2: Install marker-pdf in the virtual environment
def install_marker():
    try:
        print("Installing marker-pdf package in virtual environment...")
        subprocess.run([venv_python, "-m", "pip", "install", "marker-pdf"], check=True)
        print("marker-pdf package installed successfully.")
    except subprocess.CalledProcessError as e:
        print("Error installing marker-pdf package:", e)

# Step 3: Run marker_single for each file from names.txt
def process_files():
    try:
        # Read file names from names.txt
        with open(names_file, "r") as file:
            file_names = [line.strip() for line in file if line.strip()]

        for pdf_name in file_names:
            input_pdf_path = os.path.join(base_dir, pdf_name)
            
            # Check if the PDF file exists
            if not os.path.exists(input_pdf_path):
                print(f"Error: File not found - {input_pdf_path}")
                continue

            # Construct the marker_single command
            if os.path.exists(marker_single_exe):
                command = [marker_single_exe, input_pdf_path, "--output_dir", base_dir]
            else:
                command = [venv_python, "-m", "marker.scripts.convert_single", input_pdf_path, "--output_dir", base_dir]

            print(f"Running command: {' '.join(command)}")

            # Run the command
            result = subprocess.run(command, capture_output=True, text=True, cwd=base_dir)

            # Debugging: Log the command's outputs
            print("Command Output:", result.stdout)
            print("Command Error:", result.stderr)

#             # Save output and errors to log files in the base directory
#             with open(os.path.join(base_dir, f"{pdf_name}_output.log"), "w") as log_file:
#                 log_file.write(result.stdout)
#             with open(os.path.join(base_dir, f"{pdf_name}_error.log"), "w") as error_file:
#                 error_file.write(result.stderr)

            if result.returncode == 0:
                print(f"Processed {pdf_name} successfully. Outputs saved.")
            else:
                print(f"Error processing {pdf_name}. Check {pdf_name}_error.log for details.")

    except Exception as e:
        print("Error processing files:", e)

# Main execution
if __name__ == "__main__":
    try:
        create_virtual_env()
        install_marker()
        process_files()
    except Exception as error:
        print("An error occurred:", error)
```
